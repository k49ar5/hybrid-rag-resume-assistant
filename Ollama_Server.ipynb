```python
import os
import threading
import time
import subprocess
from pyngrok import ngrok
#TODO  Enter your Ngrok token here
NGROK_AUTH_TOKEN = <YOUR_TOKEN_HERE>
ngrok.set_auth_token(NGROK_AUTH_TOKEN)

def run_ollama():
    os.environ['OLLAMA_HOST'] = '0.0.0.0:11434'
    process = subprocess.Popen(['ollama', 'serve'])
    print("Ollama server starting...")

    time.sleep(5)

    print(" Pulling Llama3 model (this takes a while)...")
    subprocess.run(['ollama', 'pull', 'llama3'])
    print(" Llama3 ready!")

threading.Thread(target=run_ollama).start()

time.sleep(10)

public_url = ngrok.connect(11434).public_url

print(f"\n TWOJE PUBLICZNE API: {public_url}")

```
